```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model 
```

```python
df=pd.read_excel(r"C:\Users\nagas\OneDrive\Desktop\canada_per_capita_income.xlsx")
df
```

```python
plt.scatter(df.year,df.income,color="blue",marker="o")
plt.xlabel("Year")
plt.ylabel("per capita Income")
plt.show()
```

```python
reg_model=linear_model.LinearRegression()

reg_model.fit(df[["year"]],df[["income"]]) 
```

```python
reg_model.predict([[2020]])
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```
